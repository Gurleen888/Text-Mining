```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
f = open('tr')
raw = f.read().splitlines()
f.close()
print(raw[0])
print(raw[1])
print(type(raw))
```

```python
X_train_raw = []
Y_train_raw = []

import random
random.shuffle(raw)

for line in raw:
    X_train_raw.append(line[2:])
    Y_train_raw.append(int(line[0]))

Y_train = np.array(Y_train_raw)
print(len(X_train_raw))
print(X_train_raw)
print(len(Y_train))
print(Y_train)
print(Y_train.sum())
```

```python
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train_raw)
```

```python
print(vectorizer.get_feature_names())
print(X_train_tfidf.shape)
print(type(X_train_tfidf))
X_train = X_train_tfidf.toarray()
print(type(X_train))
print(X_train.shape)
```

```python
f = open('te')
read_test = f.read().splitlines()
f.close()

X_test_raw = []
Y_test_raw = []

random.shuffle(read_test)

for line in read_test:
    X_test_raw.append(line[2:])
    Y_test_raw.append(int(line[0]))

Y_test = np.array(Y_test_raw)
print(len(X_test_raw))
print(X_test_raw)
print(len(Y_test))
print(Y_test)
```

```python
X_test_tfidf = vectorizer.transform(X_test_raw)
print(X_test_tfidf.shape)
print(type(X_test_tfidf))
X_test = X_test_tfidf.toarray()
print(type(X_test))
print(X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
```

```python
model_parameters = {
        'n_estimators':[50, 100, 150, 200],
        'criterion':['gini', 'entropy'],
        'max_depth': [3, 5, 7, 9, 11]
    }
```

```python
model = RandomForestClassifier(random_state=1)
gscv = GridSearchCV(estimator=model, 
                    param_grid=model_parameters, 
                    cv=5, 
                    verbose=1, 
                    n_jobs=-1,
                    scoring='roc_auc')

gscv.fit(X_train, Y_train)
```

```python
print('The best parameter are -', gscv.best_params_)
```

```python
print(gscv.best_score_)
print(gscv.best_estimator_)
print(gscv.scorer_)
```

```python
print('AUC on test by gscv =', roc_auc_score(y_true=Y_test,
                                                        y_score=gscv.predict_proba(X_test)[:, 1]))
```
