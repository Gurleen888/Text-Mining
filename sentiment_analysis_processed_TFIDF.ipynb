```python
text = '' # text is an empty string

for i in raw:
    text += re.sub('[^a-z ]', '', i.strip()[2:].lower())+' '
    # text = text + re.sub('[^a-z ]', '', i.strip()[2:].lower())+' '

# i.strip() removes extra spaces from the start and end of a line of text
#   starting from   here and ending here  
# i.strip()[2:] # taking the text part of each line
# i.strip()[2:].lower() # Any capital letter will become smalls by using .lower()
# re.sub --> sub means substitute


print(text)
```

```python
nltk.word_tokenize(text) # returns individual words from the given text in the form of a list, this is tokenizer
# We are using word_tokenize function of nltk here, but alternatively we can also use .split()
```

```python
temp = [1, 2, 3, 3, 4, 2, 4, 4]
list( set(temp) ) # we are creating a list of unique items
```

```python
word = list( set( nltk.word_tokenize(text) ) ) # word is now the list of all the unique words in the 4000 lines of text
print(len(word)) # word is the list of all the unique words available in the trainid text data
```

```python
word # list of all the unique words
```

```python
filtered = list(set([w for w in word if not w in stopwords.words('english')])) # Python's single line loop
print(len(filtered))
print(filtered)
```

```python
stopwords.words('english') # list of all english stopwords obtained from nltk wordnet
# ASCII - American Standard Code for Information Interchange a-z, A-Z, :-), #, $, 0-9
# Unicode - International standard for all characters, devanagri script, gurmukhi, 
# .unicode() - uni --> universal
```

```python
temp = [] # empty list
for w in word: # word is a list of unique words
    if not w in stopwords.words('english'): # for every word in the list of 1625 unique words, we check whether it belongs to the
    # list of stopwords.If it belongs to stopwords, then we remove it otherwise we include it in temp for future use(tf-idf).
        temp.append(w) # list of all non-stopword unique words

filtered = list(set(temp)) # filtered will be the "list of all unique non-stop words"
print(len(filtered))
print(filtered)
# 1625 - 1515 = 110 stopwords were there in our list of 1625 unique words. And so 110 words were stopwords which have been removed.
```

```python
# filtered is the list of 1515 unique non-stopwords --> lemmatize to remove repetitive words like play, played to reduce this number 1515
wordnet_lemmatizer = WordNetLemmatizer()

wordnet_lemmatized_v = [] # empty list

for item in filtered: # 1515
    wordnet_lemmatized_v.append(wordnet_lemmatizer.lemmatize(item, 'v')) # v means verb, we will consider play --> play and played --> play as same 

wordnet_v = list(set(wordnet_lemmatized_v)) # wordnet_v is the new list which has unique non-stopwords and lemmatized words
print(len(wordnet_v))
```

```python
wordnet_v # final list of 1366 unique non-stopwords and lemmatized words --> 1366 becomes number of columns in tf-idf table
```

```python
# 1366 list of unique words across all the 4000 lines --> 1366 number of columns in tf-idf table
# Now we will, start processing each sentence to convert it into tf-idf row with 1366 values
# Till now we have the columns for tf-idf table ready

# Now we are trying to create the rows of the tf-idf table
train_file_wordnet_v = [] # empty list to hold the processed sentences
Y_train_raw = [] # empty list to hold the corresponding labels

# If we don't process the sentence, that it will contain an unprocessed word not available in the 1366 list of unique words
# So we will not be able to get tf-idf for that unprocessed word
# Thatswhy we process the sentence through remove repetitions using stopwords, lemmatize
# So on sentence level we don't use unique(set), because if a word occurs more than once in a sentence, it adds more meaning to the sentence

# 1 i love being a sentry for mission impossible and a station for bonkers
for line in raw: # raw was list of all unprocessed sentences, we are taking one unprocessed sentence at a time
    Y_train_raw.append(int(line[0]))
    x = re.sub('[^a-z ]', '', line.strip()[2:].lower()) # removing puntuations from the sentence
    tokens = nltk.word_tokenize(x) # list of words in the sentence
    filtered_tokens = [w for w in tokens if not w in stopwords.words('english')] # filtered_tokens is a list containing all the non-stopwords (no unique concept here)
    stemmed_tokens = ''
    for item in filtered_tokens:
        stemmed_tokens += wordnet_lemmatizer.lemmatize(item, 'v')+' '
    train_file_wordnet_v.append(stemmed_tokens[:-1]) # stemmed tokens contains the processed words
    # train_file_wordnet_v is a list of 4000 items, 1 item per processed sentence

print(len(train_file_wordnet_v))
print(train_file_wordnet_v[1])
# Because I would like to make friends who like the same things I like, and I really like Harry Potter, so I thought that joining a community like this would be a good start.
```

```python
vectorizer = TfidfVectorizer()
```

```python
X_train_tfidf = vectorizer.fit_transform(train_file_wordnet_v) # now we convert each processed/cleaned sentence into a tf-idf row
```

```python
print(vectorizer.get_feature_names())
print(X_train_tfidf.shape)
print(type(X_train_tfidf))
X_train = X_train_tfidf.toarray()
Y_train = np.array(Y_train_raw)
print(type(X_train))
print(X_train.shape) # Now we have a table of numbers ready for model building (we are done with text domain)
```

```python
test_file_wordnet_v = []
Y_test_raw = []

f = open('te')
read_test = f.read().splitlines()
f.close()

for line in read_test:
    Y_test_raw.append(int(line[0]))
    x = re.sub('[^a-z ]', '', line.strip()[2:].lower())
    tokens = nltk.word_tokenize(x)
    filtered_tokens = [w for w in tokens if not w in stopwords.words('english')]
    stemmed_tokens = ''
    for item in filtered_tokens:
        stemmed_tokens += wordnet_lemmatizer.lemmatize(item, 'v')+' '
    test_file_wordnet_v.append(stemmed_tokens[:-1])

print(len(test_file_wordnet_v))
print(test_file_wordnet_v[4])
```

```python
X_test_tfidf = vectorizer.transform(test_file_wordnet_v) # no fit because this is test data
print(X_test_tfidf.shape)
print(type(X_test_tfidf))
X_test = X_test_tfidf.toarray()
Y_test = np.array(Y_test_raw)
print(type(X_test))
print(X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier # Ensemble model
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
```

```python
model_parameters = {
        'n_estimators':[50, 100, 150, 200],
        'criterion':['gini', 'entropy'],
        'max_depth': [3, 5, 7, 9, 11]
    }
```

```python
model = RandomForestClassifier(random_state=1)
gscv = GridSearchCV(estimator=model, 
                    param_grid=model_parameters, 
                    cv=5, 
                    verbose=1, 
                    n_jobs=-1,
                    scoring='roc_auc')

gscv.fit(X_train, Y_train) # pandas dataframe or numpy.ndarray, does not accept list
```

```python
print('The best parameter are -', gscv.best_params_)
```

```python
print(gscv.best_score_)
print(gscv.best_estimator_)
print(gscv.scorer_)
```

```python
print('AUC on test by gscv =', roc_auc_score(y_true=Y_test,
                                                        y_score=gscv.predict_proba(X_test)[:, 1]))
```
